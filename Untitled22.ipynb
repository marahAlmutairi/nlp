```python
import re
from typing import List
class ArabicPreprocessor:
    def preprocess(self, text: str, tokenize: bool = True):
        text = text.lower()

        text = re.sub(r'http\S+|@\S+', '', text)
        text = re.sub(r'#', '', text)
        text = re.sub(r'[\u064B-\u065F]', '', text)
        text = re.sub(r'[أإآٱ]', 'ا', text)
        text = re.sub(r'ة', 'ه', text)
        text = re.sub(r'[^\w\s]', '', text)
        text = re.sub(r'(.)\1+', r'\1', text)
        text = text.strip()
        tokens = text.split()
        return tokens if tokenize else " ".join(tokens)

# تجربة الكود 
processor = ArabicPreprocessor()
sample = "@user   احب البرمجه!!  ؟ # "
result = processor.preprocess(sample, tokenize=False)
print(f"النتيجة: {result}")
```

```python
from collections import Counter, defaultdict
import math

class NgramLanguageModel:
    def __init__(self, n: int = 2):
        self.n = n
        self.counts = Counter()
        self.context_counts = Counter()
        self.vocab = set()

    def train(self, corpus: List[List[str]]):
        for tokens in corpus:
            padded = ['<s>'] * (self.n - 1) + tokens + ['</s>']
            self.vocab.update(padded)
            for i in range(len(padded) - self.n + 1):
                ngram = tuple(padded[i:i+self.n])
                context = ngram[:-1]
                self.counts[ngram] += 1
                self.context_counts[context] += 1

    def log_probability(self, ngram: tuple) -> float:
        # تطبيق Laplace Smoothing (Add-1) من NB_theory.pdf   و bigram_ex.pdf 
        count = self.counts[ngram] + 1
        context_count = self.context_counts[ngram[:-1]] + len(self.vocab)
        return math.log2(count / context_count)

    def perplexity(self, corpus: List[List[str]]) -> float:

        log_prob_sum = 0
        total_words = 0
        for tokens in corpus:
            padded = ['<s>'] * (self.n - 1) + tokens + ['</s>']
            for i in range(len(padded) - self.n + 1):
                log_prob_sum += self.log_probability(tuple(padded[i:i+self.n]))
                total_words += 1
        avg_log_prob = log_prob_sum / total_words
        return math.pow(2, -avg_log_prob)
```

```python
class NaiveBayesClassifier:
    def __init__(self, k: float = 1.0):
        self.k = k 
        self.log_priors = {}
        self.word_log_likelihoods = defaultdict(lambda: defaultdict(float))
        self.vocab = set()

    def train(self, documents: List[List[str]], labels: List[str]):
        n_docs = len(labels)
        label_counts = Counter(labels)

        for label, count in label_counts.items():
            self.log_priors[label] = math.log(count / n_docs)

        word_counts_per_class = defaultdict(Counter)
        total_words_per_class = Counter()

        for doc, label in zip(documents, labels):
            for word in doc:
                word_counts_per_class[label][word] += 1
                total_words_per_class[label] += 1
                self.vocab.add(word)

        for label in label_counts:
            denom = total_words_per_class[label] + len(self.vocab)
            for word in self.vocab:
                count = word_counts_per_class[label][word]
                self.word_log_likelihoods[label][word] = math.log((count + self.k) / denom)

    def predict_one(self, tokens: List[str]) -> str:
        scores = {}
        for label in self.log_priors:
            score = self.log_priors[label]
            for word in tokens:
                if word in self.vocab:
                    score += self.word_log_likelihoods[label][word]
            scores[label] = score
        return max(scores, key=scores.get)
```

```python
import pandas as pd
import random

df = pd.read_excel('arabic_sentiment.xlsx')

df.columns = df.columns.str.strip().str.lower()

print("الأعمدة الموجودة:", df.columns)

TEXT_COL = None
LABEL_COL = None

for col in df.columns:
    sample_values = df[col].astype(str).str.lower().head(10)

    if sample_values.str.contains(r'[ء-ي]').any():
        if TEXT_COL is None:
            TEXT_COL = col

    if sample_values.isin(['positive', 'negative']).any():
        LABEL_COL = col

if TEXT_COL is None:
    TEXT_COL = df.columns[0]

if LABEL_COL is None:
    LABEL_COL = df.columns[1]

print(f"تم اختيار عمود النص: {TEXT_COL}")
print(f"تم اختيار عمود الليبل: {LABEL_COL}")

preprocessor = ArabicPreprocessor()
nb = NaiveBayesClassifier()

train_size = int(len(df) * 0.7)
train_df = df.head(train_size)
test_df = df.tail(len(df) - train_size)

print("\nجاري تدريب الموديل...")


train_docs = [
    preprocessor.preprocess(str(t))
    for t in train_df[TEXT_COL]
]

train_labels = (
    train_df[LABEL_COL]
    .astype(str)
    .str.lower()
    .tolist()
)

nb.train(train_docs, train_labels)


print("\n" + "=" * 80)
print(f"{'التغريدة النظيفة':<45} | {'التوقع':<12} | {'الحقيقة':<10}")
print("-" * 80)

correct = 0

random.seed(42)
test_sample = test_df.sample(min(50, len(test_df)))

for i, (_, row) in enumerate(test_sample.iterrows()):

    tokens = preprocessor.preprocess(str(row[TEXT_COL]))

    prediction = nb.predict_one(tokens)

    actual = str(row[LABEL_COL]).lower()

    if prediction == actual:
        correct += 1

    if i < 15:
        clean_text = " ".join(tokens)[:43]
        print(f"{clean_text:<45} | {prediction:<12} | {actual:<10}")

accuracy = (correct / len(test_sample)) * 100

print("-" * 80)
print(f"تم فحص {len(test_sample)} تغريدة.")
print(f"الدقة النهائية: {accuracy:.2f}%")
```
